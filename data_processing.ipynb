```python
df = pd.read_csv(str(path) +'/data.csv',encoding= 'unicode_escape')
df
```

## DATA PROCESSING

```python
df = df[["CustomerID","Description","Quantity","UnitPrice","Country"]]
df.dropna()
df
```

```python
description_dictionary = {}
country_dictionary = {}
data = df.to_numpy()
num_of_rows = data.shape[0]
description_count = 1
country_count = 1

for i in range(num_of_rows):
    description = data[i][1]
    country = data[i][4]
    
    if description not in description_dictionary:
        description_dictionary[description] = description_count
        description_count = description_count + 1
    
    if country not in country_dictionary:
        country_dictionary[country] = country_count
        country_count = country_count + 1
    
print(len(country_dictionary))
print(len(description_dictionary))
```

```python
for i in range(num_of_rows):
    data[i][1] = description_dictionary[data[i][1]]
    data[i][4] = country_dictionary[data[i][4]]

print(data)
```

```python
df = pd.DataFrame(data, columns=["CustomerID","ItemID","Quantity","UnitPrice","CountryID"])
df
```

```python
df.to_csv("processed_data.csv")
```
